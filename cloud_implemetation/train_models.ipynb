# Train and Validate Severity Estimation model

```python
%cd datasets/Severity_Estimation
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
model.train(data="data.yaml", epochs=25, device = 0)
```

```python
model = YOLO("runs/detect/train2/weights/best.pt")

metrics = model.val(data="data.yaml", split="test")

metrics.results_dict
```

# Train and Validate Vehicle Detection model wit imgsz=150
